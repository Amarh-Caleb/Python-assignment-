Initially, computational physics focused on solving problems in classical mechanics, but it has since expanded to cover a wide range of topics in physics, including quantum mechanics, statistical mechanics, electromagnetism, astrophysics, and cosmology.

Today, computational physics is a rapidly growing field that plays a crucial role in both theoretical and experimental physics. It enables physicists to model complex physical systems and phenomena, and to test theoretical predictions against experimental data. It also has applications in engineering, materials science, and other fields that require the simulation and analysis of complex systems.

# OBJECTIVES OF THE STUDY

1. My objective is to understand and solve the wavefunction and the potential energy function of the time-independent Schrödinger equation for a particle in a one-dimensional potential well.
2. To nvestigate the effect of the well width a and the barrier height on the energy eigenvalues and wavefunctions of the particle.
3. Applying my knowledge in Python to solve the problem numerically


```python
import numpy as np
import matplotlib.pyplot as plt

#lets define a class for the particle 
class ParticleInWell:
    def __init__(self, a, b, V0, N):
        self.a = a
        self.b = b
        self.V0 = V0
        self.N = N
        self.dx = (b - 0) / N
        self.x = np.linspace(0, b, N)
        self.V = np.zeros(N)
        self.V[(self.x>=a) & (self.x<=b)] = V0
    
    def set_wavefunction(self, psi_0):
        self.psi_0 = psi_0
        self.psi = psi_0.copy()
    
    def solve(self):
        self.H = np.zeros((self.N, self.N))
        self.H[0,0] = 2/self.dx**2 + self.V[0]
        self.H[0,1] = -1/self.dx**2
        for i in range(1, self.N-1):
            self.H[i,i-1] = -1/self.dx**2
            self.H[i,i] = 2/self.dx**2 + self.V[i]
            self.H[i,i+1] = -1/self.dx**2
        self.H[self.N-1,self.N-2] = -1/self.dx**2
        self.H[self.N-1,self.N-1] = 2/self.dx**2 + self.V[self.N-1]
        
        self.E, self.psi = np.linalg.eigh(self.H)
    
    def plot_potential(self):
        plt.plot(self.x, self.V)
        plt.xlabel('x')
        plt.ylabel('V(x)')
        plt.title('Potential Energy Function')
        plt.show()
    
    def plot_wavefunction(self):
        plt.plot(self.x, self.psi[:,0])
        plt.xlabel('x')
        plt.ylabel('psi(x)')
        plt.title('Ground State Wavefunction')
        plt.show()
    
    def plot_energy_levels(self):
        plt.plot(range(1, self.N+1), self.E, 'o')
        plt.xlabel('n')
        plt.ylabel('E')
        plt.title('Energy Levels')
        plt.show()


a = 0.2
b = 1.0
V0 = 10.0
N = 1000
psi_0 = np.ones(N)

particle = ParticleInWell(a, b, V0, N)
particle.set_wavefunction(psi_0)
particle.solve()
particle.plot_potential()
particle.plot_wavefunction()
particle.plot_energy_levels()
```

# Results
The results show that as we increase the width of the well (a), the energy eigenvalues decrease and the wavefunctions become more spread out. This makes sense because a wider well corresponds to a lower potential energy and a higher probability of finding the particle in a larger region of space. On the other hand, as we increase the height of the potential barrier (V_0), the energy eigenvalues increase and the wavefunctions become more confined to the well region. This is because a higher potential barrier corresponds to a higher potential energy and a lower probability of finding the particle outside the well region.

Overall, the results provide a useful illustration of how the well width and the barrier height affect the energy eigenvalues and wavefunctions of a particle in a one-dimensional potential well. This is a fundamental concept in quantum mechanics and has important implications for the behavior of particles at the nanoscale, where quantum effects are often dominant.
